# Handling Null Values

```python
import pandas as pd
# csv file from kaggle 10 months past data
df = pd.read_csv("weather_data.csv",parse_dates=['day'])
df
```

```python
# shape gives (rows,columns)
df.shape
```

```python
# let's check if there is any null 
df.info()
```

```python
# Checking Null or not using isnull() method:
df.isnull()
```

```python
# It's not that much clear is it?
# so lets make it clear using sum()
df.isnull().sum()
```

```python
# total num values
df.isnull().sum().sum()
```

# Updating the csv file and handling missing values

```python
# Let's set day as index
df.set_index('day',inplace=True)
df
```

<h3>Filling values in Null using method: fillna()</h3>

```python
a = df.fillna(0)
a
```

```python
# lets change a specific values in file using dictionary in fillna

a = df.fillna({
    'temperature':5,
    'windspeed':6,
    'event':'no event'
})
a
```

<h3>using fillna(method="ffill") or fillna(method="bfill")</h3>
<h4>ffill: fills the forward data into NAN</h4>
<h4>bfill: fills the backward data into NAN</h4>

```python
# forwardfill
a = df.fillna(method="ffill")
a
```

<h3>Interpolate() to do interpolation in dataframe</h3>

```python
a = df.interpolate(method="time")
a
```

<h3>dropna method</h3>

```python
# Drops all the rows that has atleas one null value
a = df.dropna()
a
```

```python
a = df.dropna(how="all")
a
```

```python
a = df.dropna(thresh=2)
a
```

```python
# lets fill the remaining dates using method: date_range()
dt = pd.date_range('01-01-2017','01-11-2017')
index = pd.DatetimeIndex(dt)
df = df.reindex(index)
df
```
