<h3>Handling missing data using replace function.</h3>

```python
import pandas as pd
import numpy as np
```

```python
# read csv file
d = pd.read_csv("weather_data.csv")
d
```

<h3>Using replace method/function:</h3>

```python
e = d.replace([-99999,0],np.NaN)
e
```

```python
# windspeed can be zero so let's use dictionary
e = d.replace({
    "temperature":-99999,
    "windspeed":-99999,
},np.NaN)
e
```

```python
# replace specific value
f = e.replace({
    -99999: np.NaN,
    '0': 'Hot'
})
f
```

<h2>Replace characters from the table using regex:</h2>

```python
f = e.replace('[A-Za-z]',' ',regex=True)
f
```

```python
# Here we don't want to replace the events
f = e.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]',
},'',regex=True)
f
```
